### Data Visualisation With Seaborn

#### Seaborn is a library for making statistical graphics in Python. It is built on top of matplotlib and closely integrated with pandas data structures. 
Seaborn is open source and is hosted on github, the official documentation page can be found on github, to find the seaborn documentation, just search it on google. Or Visit https://seaborn.pydata.org/


```python
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#distribution plot helps to show a distribution of a particular feature of a dataset.
#It dosent make comparison of two features in a dataset.

sns.distplot(tips['size'])
#distplot can only be used on numerical data
```

```python
sns.distplot(tips['size'], kde= False, bins=50) #kde- kernel density estimator, bins- number of distribution
```

```python
sns.distplot(tips['total_bill'], kde= False, bins=50)
```

```python
sns.jointplot(x= "total_bill", y= "tip", data= tips)
# joint plot is used to make comparison betweeen two features
```

```python
sns.jointplot(x= "total_bill", y= "tip", data= tips, kind ="hex")
```

```python
sns.jointplot(x= "total_bill", y= "tip", data= tips, kind ="scatter")
```

```python
sns.jointplot(x= "total_bill", y= "tip", data= tips, kind ="kde")
```

```python
sns.pairplot(tips)

#plots every single combination of the numerical columns in the dataframe
#plots a histogram when it wants to plot a variable against itself
# The "hue" arguments allows you to pass in any categorical (non-numerical) column
#The blue points are male while the orange ones are females
#The pallete argument allows us to change the colours of our data the way we want it
```

```python
sns.pairplot(tips, hue = "sex", )
```

```python
sns.pairplot(tips, hue = "day")
```

```python
sns.rugplot(tips["total_bill"])
```

### Box and Whiskerplot
This type of plot is used to show comparison between qualitative data

```python
sns.boxplot(x="day", y="total_bill", data=tips)
#the dots outside are basically outliers
#the box shows the plot while the whiskers extend to show the rest of the distribution
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")
#Smokers pay more of thier total bil on sundays
```

```python
### Violin Plot A type of categorical plot to visualise categorical data
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color="white")
```

```python
sns.factorplot(x='day', y="total_bill", data=tips, kind="bar")
```

```python
sns.factorplot(x='day', y="total_bill", data=tips, kind="violin")
```

```python
#Heatmaps are primary ways of showing matrix maps, also used to visualize the correlation
tc = tips.corr()
tc
#tips and total_bill have the strongest corelation
```

```python
sns.heatmap(tc, annot=True, cmap="coolwarm") #annotaion is the colours
```

### Matrix Plots

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
fCr = flights.corr()
fCr
```

```python
sns.heatmap(fCr, annot=True, cmap="coolwarm") 
```

```python
fp =flights.pivot_table(index="month", columns="year", values="passengers")
fp
```

```python

sns.heatmap(fp)
#highest distribution of passengers = July 1960
```

```python
sns.heatmap(fp,cmap="coolwarm", linecolor="white", linewidths="4")
```

```python
sns.clustermap(fp)# groups rows and columns together based on similar features
```

```python
sns.clustermap(fp, cmap="coolwarm", standard_scale=1)
```

### Regression plot

```python
sns.lmplot(x='total_bill', y='tip', data= tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips, hue="sex", markers=["o", 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips, col="sex")
```

```python
sns.set_style("darkgrid", {"axis.facecolor": ".9"})
sns.lmplot(x='total_bill', y='tip', data= tips, col="sex", row='time')
```
