```python
import numpy as np

# Challenge 1.1: Create a 1D array of numbers from 0 to 9.
array_1d = np.arange(10)
print("1D array from 0 to 9:", array_1d)

# Challenge 1.2: Create a 2D array of shape (3, 3) with numbers from 1 to 9.
array_2d = np.arange(1, 10).reshape((3, 3))
print("2D array of shape (3, 3):\n", array_2d)

# Challenge 1.3: Create a 3D array of shape (2, 2, 2) with random numbers.
array_3d = np.random.random((2, 2, 2))
print("3D array of shape (2, 2, 2):\n", array_3d)

# Challenge 2.1: Add two arrays.
array_a = np.array([1, 2, 3])
array_b = np.array([4, 5, 6])
array_sum = array_a + array_b
print("Sum of arrays:", array_sum)

# Challenge 2.2: Multiply two arrays element-wise.
array_product = array_a * array_b
print("Element-wise product of arrays:", array_product)

# Challenge 2.3: Subtract one array from another.
array_subtract = array_a - array_b
print("Subtraction of arrays:", array_subtract)

# Challenge 2.4: Divide one array by another.
array_divide = array_a / array_b
print("Division of arrays:", array_divide)

# Challenge 3.1: Extract the second row from the 2D array created in 1.2.
second_row = array_2d[1, :]
print("Second row:", second_row)

# Challenge 3.2: Extract all elements from the 2nd column.
second_column = array_2d[:, 1]
print("Second column:", second_column)

# Challenge 3.3: Extract a subarray from the 3D array created in 1.3.
subarray_3d = array_3d[0, :, :]
print("Subarray from 3D array:\n", subarray_3d)

# Challenge 3.4: Reverse the 1D array created in 1.1.
reversed_array_1d = array_1d[::-1]
print("Reversed 1D array:", reversed_array_1d)

# Challenge 4.1: Calculate the mean of all elements in the 2D array.
mean_value = np.mean(array_2d)
print("Mean of 2D array:", mean_value)

# Challenge 4.2: Find the maximum value in each column.
max_in_columns = np.max(array_2d, axis=0)
print("Maximum values in each column:", max_in_columns)

# Challenge 4.3: Calculate the standard deviation of the 1D array.
std_deviation = np.std(array_1d)
print("Standard deviation of 1D array:", std_deviation)

# Challenge 4.4: Calculate the variance of the 2D array.
variance_value = np.var(array_2d)
print("Variance of 2D array:", variance_value)

# Challenge 5.1: Add a 1D array to each row of the 2D array.
array_to_add = np.array([10, 20, 30])
broadcast_sum = array_2d + array_to_add[:, np.newaxis]
print("Broadcasted sum:\n", broadcast_sum)

# Challenge 5.2: Multiply a 1D array with each column of the 2D array.
broadcast_product = array_2d * array_to_add[:, np.newaxis]
print("Broadcasted product:\n", broadcast_product)

# Challenge 6.1: Compute the dot product of two matrices.
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])
dot_product = np.dot(matrix_a, matrix_b)
print("Dot product of matrices:\n", dot_product)

# Challenge 6.2: Compute the inverse of a matrix.
matrix_c = np.array([[1, 2], [3, 4]])
inverse_c = np.linalg.inv(matrix_c)
print("Inverse of matrix:\n", inverse_c)

# Challenge 6.3: Calculate the determinant of a matrix.
determinant_c = np.linalg.det(matrix_c)
print("Determinant of matrix:", determinant_c)

# Challenge 6.4: Compute the eigenvalues of a matrix.
eigenvalues_c = np.linalg.eigvals(matrix_c)
print("Eigenvalues of matrix:", eigenvalues_c)

# Challenge 7.1: Select elements from an array using a boolean mask.
mask = array_1d % 2 == 0
selected_elements = array_1d[mask]
print("Elements selected using boolean mask:", selected_elements)

# Challenge 7.2: Select elements based on a condition.
conditioned_elements = array_2d[array_2d > 5]
print("Elements greater than 5:", conditioned_elements)

# Challenge 7.3: Index with an array of indices.
indices = np.array([0, 2, 4])
indexed_elements = array_1d[indices]
print("Elements indexed with an array of indices:", indexed_elements)

# Challenge 7.4: Use fancy indexing to assign new values.
array_1d[indices] = 100
print("Array after fancy indexing assignment:", array_1d)

# Challenge 8.1: Generate a 1D array of 10 random numbers between 0 and 1.
random_1d = np.random.rand(10)
print("1D array of random numbers:", random_1d)

# Challenge 8.2: Generate a 2D array of shape (3, 3) with random integers between 1 and 10.
random_2d = np.random.randint(1, 10, size=(3, 3))
print("2D array of random integers:\n", random_2d)

# Challenge 8.3: Generate a random sample from a normal distribution.
random_normal = np.random.normal(size=(3, 3))
print("Random sample from normal distribution:\n", random_normal)

# Challenge 8.4: Shuffle a 1D array in place.
np.random.shuffle(array_1d)
print("Shuffled 1D array:", array_1d)

# Challenge 9.1: Reshape the 1D array to a 2D array of shape (2, 5).
reshaped_array = array_1d.reshape((2, 5))
print("Reshaped array (2, 5):\n", reshaped_array)

# Challenge 9.2: Flatten the 2D array created in 1.2.
flattened_array = array_2d.flatten()
print("Flattened array:", flattened_array)

# Challenge 9.3: Flatten the 3D array created in 1.3.
flattened_3d_array = array_3d.flatten()
print("Flattened 3D array:", flattened_3d_array)

# Challenge 9.4: Reshape the 2D array created in 1.2 to (1, 9).
reshaped_2d_array = array_2d.reshape((1, 9))
print("Reshaped 2D array (1, 9):\n", reshaped_2d_array)

# Challenge 10.1: Calculate the cumulative sum of the 1D array created in 1.1.
cumulative_sum = np.cumsum(array_1d)
print("Cumulative sum of 1D array:", cumulative_sum)

# Challenge 10.2: Calculate the cumulative product of the 1D array created in 1.1.
cumulative_product = np.cumprod(array_1d)
print("Cumulative product of 1D array:", cumulative_product)

# Challenge 10.3: Find the unique elements in the 1D array created in 1.1.
unique_elements = np.unique(array_1d)
print("Unique elements in 1D array:", unique_elements)

# Challenge 10.4: Find the indices of the unique elements in the 1D array created in 1.1.
unique_indices = np.unique(array_1d, return_index=True)
print("Indices of unique elements in 1D array:", unique_indices)
```
